## Introduction

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's(another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures.

Some of the major Pros of Matplotlib are:

* Generally easy t oget started for simple plots.
* Support for custom labels and texts.
* Great control of every element in a figure.
* High - quality output in many formats
* Very customizable in general

Matplotlib allows you to crate repreducible figures programmatically . Let's learn how to use it!

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x,y)   # if want to give colour as example red plt.plot(x,y,'r-')
# plt.show(),

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) 
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
# This use to draw one plot in another plot.

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.30])  #(x,y,height,width)

# axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

  ## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=0)  # this use to show info box in graph
          

# ax.set_title('Title')
# ax.set_ylabel('Y')
# ax.set_xlabel('X')
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
# This command is use to save the figure
fig.savefig('my_picture.png',dpi=200)
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,alpha=0.5,ls =':',marker='*',markersize=25,
       markerfacecolor='yellow',markeredgewidth=2) # we can also use RBG Hex Code for colors.
# alpha increases the accuracy of the line
# linewidth   = it increase the width of the line. also use "lw"

# linestyle(ls)= ':','--','-.','steps',('-')=> this means only plain line 
# marker show any structure on perticular point
```

```python
x
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color='purple',lw=2,ls='--')

axes.set_xlim([0,1])
axes.set_ylim([0,2])
```

## Hostogram

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('First Plot')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.plot(x,z)

ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoom')
ax2.plot(x,y)
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```
