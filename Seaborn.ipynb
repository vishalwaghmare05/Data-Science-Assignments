```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup Completed')
```

```python
dt = pd.read_csv('fifa.csv', index_col='Date',parse_dates=True)
# This tells the notebook to understand the each row label as a date (as opposed to a number or other text with a different meaning).
```

```python
dt.head()
```

# Line Plot

```python
# set the width and height of the figure.
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time.
sns.lineplot(data=dt)
```

```python
dt.columns
```

```python
plt.figure(figsize=(14,6))
plt.title('fifa data')
sns.lineplot(data=dt['ARG'],label=("ARG"))
sns.lineplot(data=dt.BRA,label=('BRA'))
plt.xlabel("Date")
```

```python
data = pd.read_csv('flight_delays.csv',index_col='Month')
data.head()
```

```python
# change the position of the labels in this graph.
plt.figure(figsize=(14,6))
sns.lineplot(data=data).legend(loc='upper right')
```

# Bar Chart

```python
plt.figure(figsize=(10,6))

plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airline fights by months
sns.barplot(x=data.index,y=data['NK'])

plt.ylabel("Arrival Delay (in minutes)")
```

# Heatmap

```python
plt.figure(figsize=(14,7))

plt.title("Average Arrival Delay for each Airline, by Month")

sns.heatmap(data=data, annot=True)
# This ensures that the values for each cell appear on the chart. (Leaving this out removes the numbers from each of the cells!)


plt.xlabel('Airline')
```

# Scatter Plot

```python
data = pd.read_csv('insurance.csv')
data.head()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x=data['bmi'],y=data['charges'])
```

```python
# we can add best fitted regression line.
plt.figure(figsize=(12,7))
sns.regplot(x=data['bmi'],y=data['charges'],color='red')
```

The hue parameter in a Seaborn bar plot allows you to add an additional categorical variable to your plot. It helps to differentiate the bars by a specific category, enabling you to compare multiple variables simultaneously.

```python
# Color - coded scatter plots
plt.figure(figsize=(12,7))
sns.scatterplot(x=data['bmi'],y=data['charges'],hue = data['smoker'])
```

```python
# for adding two regression lines for smokers and nonsmokers.
plt.figure(figsize = (14,6))
sns.lmplot(x="bmi",y="charges",hue="smoker",data=data)
```

## Categorical Scatter plot   
using sns.swarmplot

```python
plt.figure(figsize=(12,8))
sns.swarmplot(x=data['smoker'],y=data['charges'])
```
